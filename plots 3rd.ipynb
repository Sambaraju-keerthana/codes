```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Downloads/mpg_raw.csv")
```

```python
df.head
```

```python
df.dropna(inplace =True)
```

```python
df[['mpg','cylinders','displacement','horsepower','weight','acceleration']].hist(figsize=(18,12))

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot( df, x='weight', y='mpg')
plt.title("Scatter plot for weight versus mpg")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot( df, x='cylinders')
plt.title("Point plot for cylnders")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.pointplot( df, x='model_year', y='mpg')
plt.title("point plot for model year versus mpg")
plt.show()
```
