```python
import numpy as np

# Creating array object
arr = np.array( [[ 1, 2, 3],
                 [ 4, 2, 5]] )
print(arr)

# Printing type of arr object
print("Array is of type: ", type(arr))

# Printing array dimensions (axes)
print("No. of dimensions: ", arr.ndim)

# Printing shape of array
print("Shape of array: ", arr.shape)

# Printing size (total number of elements) of array
print("Size of array: ", arr.size)

# Printing type of elements in array
print("Array stores elements of type: ", arr.dtype)
```

```python
import numpy as np
# create a 2D array
array1 = np.array([[1, 3], [5, 7]])
# calculate the mean of the entire array
result1 = np.mean(array1)
print("Entire Array:",result1)  # 4.0
```

```python
# calculate the mean along vertical axis (axis=0)
result2 = np.mean(array1, axis=0)
print("Along Vertical Axis:",result2)  # [3. 5.]

# calculate the mean along  horizontal axis (axis=1)
result3 = np.mean(array1, axis=1)
print("Along Horizontal Axis :",result3)  # [2. 6.]
```

```python
# Reshaping 3X4 array to 2X2X3 array

import numpy as np
arr = np.array([[1, 2, 3, 4],
                [5, 2, 4, 2],
                [1, 2, 0, 1]])

newarr = arr.reshape(2,2,3)

print ("Original array:\n", arr)
print("---------------")
print ("Reshaped array:\n", newarr)
```

```python
# Using -1 to infer a dimension
# Reshaping the array into a 2D array
# '-1' allows to calculate the number of rows based on the total number of 
#elements
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped_arr = np.reshape(arr, (-1, 2))  
print(reshaped_arr)
''' order : [C-contiguous, F-contiguous, A-contiguous; optional]

    'C' (default): Row-major order.
    'F': Column-major order.
    'A': Fortran-like index order if the array is Fortran-contiguous; otherwise, C-like order.
    'K': Keeps the array's order as close to its original as possible.'''
# Reshaping the array into a 2D array with 2 rows and 3 columns 
# with column-major order
reshaped_arr = np.reshape(arr, (2, 3), order='F')
print(reshaped_arr)
```

```python
# Flatten array
arr = np.array([[1, 2, 3], [4, 5, 6]])
flat_arr = arr.flatten()

print ("Original array:\n", arr)
print ("Fattened array:\n", flat_arr)
```

```python
''' A DataFrame is a two-dimensional labeled data structure with columns of 
potentially different types. It can be thought of as a table or a spreadsheet,
consisting of rows and columns. Each column in a DataFrame is a Series 
object, sharing a common index for the rows.
DataFrames are highly versatile and are the primary structure for working 
with tabular data in pandas.'''
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 32, 18, 47, 33],
    'City': ['New York', 'Paris', 'London', 'Tokyo', 'Sydney']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# access a single row
single_row = df.loc[2]
print("Single row:", single_row)
print()

# access rows 0, 3 and 4
row_list = df.loc[[0, 3, 4]]
print("List of Rows:", row_list)
print()


# access a list of columns
column_list = df.loc[:, ['Name', 'Age']]
print("List of Columns:", column_list)
print()


# access second row of 'Name' column
specific_value = df.loc[1, 'Name']
print("Specific Value:", specific_value)
```

```python
#Adding Rows Using Concat()
#create the first data frame
import pandas as pd
data= {'Name': ['John', 'Henry', 'Jack', 'Mary'],
      'Age':[23, 34, 25, 27], 
      "Address":['Bengaluru', 'Hubli', 'Hydrabad', 'Pune']}
df1=pd.DataFrame(data)

#create the second data frame
data1= {'Name': ['Jimmy', 'Herry'],
      'Age':[33, 34], 
      "Qualification":['MCA', 'BE']}
df2=pd.DataFrame(data1)

df3=pd.concat([df1,df2], axis=1)
df3
```

```python
#droping rows by index label
import pandas as pd
data= {'Name': ['John', 'Henry', 'Jack', 'Mary'],
      'Age':[23, 34, 25, 27], 
      "Address":['Bengaluru', 'Hubli', 'Hydrabad', 'Pune']}
df1=pd.DataFrame(data)
# Setting the index to 'Name'
df1.set_index('Name', inplace=True)
# dropping the rows mentioned
df1.drop(['Henry'], inplace=True)
df1
```

```python
# Sorting the Data farme, here na_position= 'last' moves 
#the rows contatining the NA values to the last and sort the remaining rows
df1.sort_values("Name", axis=0, inplace=True, ascending = False, na_position='last');
df1
```

```python
#Grouping the data
# groupby() method can be used. Python allows to compute aggreagations such as sum(), mean(), median()
#min(), max() on the gropued data
import pandas as pd
data= {'Name': ['John', 'Henry', 'Jack', 'Mary', 'Mary', 'Joseph'],
      'Age':[23, 34, 25, 27, 32, 40], 
      'Address':['Bengaluru', 'Hubli', 'Hydrabad', 'Pune', 'Bengaluru', 'Pune'],
      'Salary': [25000, 32000, 56000, 100000, 45000, 120000]}
df1=pd.DataFrame(data)
#calculating the mean salary grouped by address
age_group=df1['Age'].groupby(df1['Address'])

print(age_group.mean())
```

```python
'''
Pandas Series:

    A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, Python objects, etc.).
    It is essentially a single column of data, similar to a NumPy array but with an associated index that labels each element.
    The index can be numerical (default) or custom labels (e.g., strings).
    While the data within a Series is mutable (can be changed), the size of the Series itself is immutable (cannot be changed after creation).
```

```python
import pandas as pd 
import numpy as np
# Creating empty series 
ser = pd.Series() 
print("Pandas Series: ", ser) 

# simple array 
data = np.array(['P', 'G', 'M', 'C', 'A'])   
ser = pd.Series(data) 
print("Pandas Series:\n", ser)
```

```python
'''Creating a series from array with an index: In order to create a series by explicitly providing 
index instead of the default, we have to provide a list of elements to the index parameter with 
the same number of elements as it is an array. '''
# import pandas as pd
import pandas as pd

# import numpy as np
import numpy as np

# simple array
data = np.array(['P', 'G', 'M', 'C', 'A'])

# providing an index
ser = pd.Series(data, index=[10, 11, 12, 13, 14])
print(ser)
```

```python
'''Creating a series from Lists: In order to create a series from list, 
we have to first create a list after that we can create a series from list. '''
import pandas as pd

# a simple list
list = ['P', 'G', 'M', 'C', 'A']

# create series form a list
ser = pd.Series(list)
print(ser)
```

```python
'''Creating a series from Dictionary: In order to create a series from the dictionary, 
we have to first create a dictionary after that we can make a series using dictionary. 
Dictionary keys are used to construct indexes of Series. '''
import pandas as pd

# a simple dictionary
dict = {'rvce': 10,
		'for': 20,
		'MCA': 30}

# create series from dictionary
ser = pd.Series(dict)

print(ser)
```

```python
'''Creating a series from Scalar value: In order to create a series from scalar value, 
an index must be provided. The scalar value will be repeated to match the length of the index. '''
import pandas as pd

import numpy as np

# giving a scalar value with index
ser = pd.Series(10, index=[0, 1, 2, 3, 4, 5])

print(ser)
```

```python
'''Creating a series using NumPy functions : In order to create a series using numpy function, 
we can use different function of numpy like numpy.linspace(), numpy.random.radn(). '''
# import pandas and numpy
import pandas as pd
import numpy as np

# series with numpy linspace()
ser1 = pd.Series(np.linspace(3, 33, 3))
print(ser1)

# series with numpy linspace()
ser2 = pd.Series(np.linspace(1, 100, 10))
print("\n", ser2)
```
